```python
playoffs = {
    "WildCard": 18,
    "Division": 19,
    "ConfChamp": 20,
    "SuperBowl": 21
}
```

```python
def get_int_week(week):
    try:
        return int(week)
    except ValueError:
        return playoffs[week]
```

```python
games_df["int_week"] = games_df["week"].apply(get_int_week)
```

```python
games_df.head()
```

```python
giants_home_games = games_df["home_team_name"] == "Giants"
giants_away_games = games_df["away_team_name"] == "Giants"
games_in_2007 = games_df["year"] == 2007

giants_games_2007 = games_df[(giants_home_games | giants_away_games) & games_in_2007]
giants_games_2007.sort_values("int_week")[giants_games_2007.columns[:-1]]
```

```python
giants_playoff_games_2007 = giants_games_2007[giants_games_2007["int_week"] > 17]
giants_playoff_games_2007
```

```python
super_bowls = games_df[games_df["int_week"] == playoffs["SuperBowl"]]
```

```python
super_bowls
```

```python
super_bowls["win_team_name"].value_counts()
```

```python
from collections import Counter

sb_counter = Counter()

for idx in super_bowls.index:
    row = super_bowls.loc[idx]
    sb_counter[row["win_team_name"]] += 1
    sb_counter[row["loss_team_name"]] += 1

for team, count in sb_counter.most_common():
    print(f"{team}: {count}")
```

```python
super_bowls_played = pd.Series(sb_counter).sort_values(ascending=False)
super_bowls_played
```

```python
super_bowls_won = super_bowls["win_team_name"].value_counts()
super_bowls_lost = super_bowls["loss_team_name"].value_counts()
```

```python
super_bowls_df = pd.concat([super_bowls_played, super_bowls_won, super_bowls_lost], axis=1)
super_bowls_df
```

```python
super_bowls_df.fillna(0, inplace=True) 
super_bowls_df
```

```python
super_bowls_df = super_bowls_df.astype(int)
super_bowls_df
```

```python
super_bowls_df.columns = ["played", "won", "lost"]
super_bowls_df
```

```python
super_bowls_df["pct_won"] = round(super_bowls_df["won"] / super_bowls_df["played"], 2)
```

```python
super_bowls_df
```

```python
super_bowls_df.sort_values(["pct_won", "played"], ascending=False)
```
